```python

# SISTEMA DE VERIFICAÇÃO DE IDADE PARA EVENTOS

# Regras principais do projeto:
# - Menor que 16 anos: NÃO entra.
# - 16 ou 17 anos: Entra SOMENTE com responsável/acompanhado dos pais
# - 18 anos ou mais: Entrada liberada!


# Essa primeira parte é a Entrada de dados! onde vamos
#colocar  os dados, como: nome e idade do usuário.

nome = input("Digite seu nome: ")
idade = int(input("Digite sua idade: "))

#Depois de informar e salvar os dados, o programa faz a verificação da idade.

# Verificação da idade
if idade < 16:
    print(f"{nome}, entrada NÃO permitida.")
    print("Motivo: Menor de 16 anos.")

elif idade == 16 or idade == 17:
    print(f"{nome}, entrada permitida SOMENTE com responsável.")
    print("Classificação: Menor acompanhado.")

else:
    print(f"{nome}, entrada PERMITIDA.")
    print("Classificação: Maior de idade.")

# Saída final
print("Sistema de verificação encerrado.")


# IF, ELIF E ELSE são as nossas condicioanais dentro do nosso programa.

```
