```python
alunos = {
    "Alice": 8.5,
    "Bruno": 7.8,
    "Carla": 9.2,
    "Diego": 6.9,
    "Eva": 8.0,
    "Fernando": 7.5,
    "Gabriela": 9.0,
    "Henrique": 6.7,
    "Isabela": 8.3,
    "Jo√£o": 7.2
}

# Quantos alunos existem?
print("Quantidade de alunos:", len(alunos))

# Algu√©m tirou nota zero?
if 0 in alunos.values():
    print("Sim, algu√©m tirou nota zero.")
else:
    print("Ningu√©m tirou nota zero.")

# Menor e maior nota
menor_nota = min(alunos.values())
maior_nota = max(alunos.values())

# Encontrar aluno pela nota
def buscar_aluno_por_nota(nota):
    for nome, n in alunos.items():
        if n == nota:
            return nome

print(f"A menor nota foi {menor_nota} do(a) aluno(a) {buscar_aluno_por_nota(menor_nota)}")
print(f"A maior nota foi {maior_nota} do(a) aluno(a) {buscar_aluno_por_nota(maior_nota)}")
```

```python
bandas = {
    "Gangrena Gasosa": [
        {"√°lbum": "Gente Ruim S√≥ Manda Lembran√ßa Pra Quem N√£o Presta", "ano": 2019},
        {"√°lbum": "Se Deus √â 10, Satan√°s √â 666", "ano": 1999},
        {"√°lbum": "Welcome to the Macumba", "ano": 1993}
    ],
    "Rog√©rio Skylab": [
        {"√°lbum": "Nas Portas do Cu", "ano": 2023},
        {"√°lbum": "Cr√≠tica da Faculdade do Cu", "ano": 2021},
        {"√°lbum": "Cosmos", "ano": 2019}
    ],
    "Garotos Podres": [
        {"√°lbum": "Can√ß√µes para Ninar", "ano": 2003},
        {"√°lbum": "Com a Corda Toda", "ano": 1997},
        {"√°lbum": "Rock de Sub√∫rbio", "ano": 1995}
    ],
    "Massacration": [
        {"√°lbum": "Live Metal Espancation", "ano": 2017},
        {"√°lbum": "Good Blood Headbanguers", "ano": 2009},
        {"√°lbum": "Gates of Metal Fried Chicken of Death", "ano": 2005}
    ],
    "Raimundos": [
        {"√°lbum": "Cantigas de Roda", "ano": 2014},
        {"√°lbum": "Roda Viva", "ano": 2006},
        {"√°lbum": "√âramos 4", "ano": 2001}
    ]
}


# ------------------------------
# 1) √ÅLBUM MAIS ANTIGO
# ------------------------------

album_mais_antigo = None
ano_mais_antigo = 9999
banda_mais_antiga = None

for banda, lista in bandas.items():
    for album in lista:
        if album["ano"] < ano_mais_antigo:
            ano_mais_antigo = album["ano"]
            album_mais_antigo = album["√°lbum"]
            banda_mais_antiga = banda

print(f"üìÄ √Ålbum mais antigo: '{album_mais_antigo}' ({ano_mais_antigo}) da banda {banda_mais_antiga}\n")



# ------------------------------
# 2) BANDA COM MAIOR INTERVALO ENTRE √ÅLBUNS
# ------------------------------

maior_intervalo = -1
banda_intervalo = None

for banda, lista in bandas.items():
    anos = sorted([album["ano"] for album in lista])
    intervalo = anos[-1] - anos[0]   # ano mais recente - ano mais antigo

    if intervalo > maior_intervalo:
        maior_intervalo = intervalo
        banda_intervalo = banda

print(f"‚è≥ Banda com maior intervalo entre √°lbuns: {banda_intervalo} ({maior_intervalo} anos)\n")



# ------------------------------
# 3) IMPRIMIR TODOS OS √ÅLBUNS DO MAIS NOVO PRO MAIS ANTIGO
# ------------------------------

todos_os_albuns = []

for banda, lista in bandas.items():
    for album in lista:
        todos_os_albuns.append((album["√°lbum"], album["ano"], banda))

# ordenar do mais novo para o mais antigo
todos_os_albuns.sort(key=lambda x: x[1], reverse=True)

print("üìå √Ålbuns do mais novo para o mais antigo:\n")
for nome, ano, banda in todos_os_albuns:
    print(f"{ano} - {nome}  |  {banda}")
```

```python
animes = {
    "Naruto": 220,
    "Jujutsu Kaisen": 47,
    "Dragon Ball Z": 291,
    "Death Note": 37,
    "Fullmetal Alchemist": 64,
    "Evangelion": 26,
    "Berserk": 25,
    "Code Geass": 50,
    "Akame ga Kill!": 24,
    "Elfen Lied": 13
}

# Total de epis√≥dios
total_episodios = sum(animes.values())
print("Total de epis√≥dios:", total_episodios)

# Transformar episodios em percentual
animes_percent = {
    anime: (eps / total_episodios) * 100
    for anime, eps in animes.items()
}

# Ordenar em ordem decrescente
ordenado = sorted(animes_percent.items(), key=lambda x: x[1], reverse=True)

# Impress√£o formatada
print("\n--- Percentual de epis√≥dios por anime ---\n")
for anime, pct in ordenado:
    print(f"Anime: {anime:<20} | Percentual: {pct:5.2f}%")
```
